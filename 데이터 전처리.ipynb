# 결측치 제거

```python
erase_df = nan_df # 복사본 
```

erase_df : 결측치가 제거된 데이터 프레임

```python
erase_df.dropna()

erase_df
```

inplace=True : 원본에 반영을 하겠다는 뜻

```python
erase_df.dropna(inplace = True)

erase_df
```

# 미리보기 column : 한글만 포함, 여러 공백 제거, 짧은거 제거

```python
preview_df = erase_df # 복사본
```

preview_df : 미리보기를 불용어 제거한 데이터 프레임

```python
preview_df.preview = preview_df.preview.str.replace(r"[^가-힣 \s]", " ", regex=True) # 가 각 간 부터 힢 힡 힣 빼고는 다 공백
preview_df.preview = preview_df.preview.str.replace(r"\s+", " ", regex=True) # 연속한 공백을 하나의 공백으로
preview_df.preview = preview_df.preview.str.strip() # 양쪽에 공백 있으면 제거됨

preview_df = preview_df[preview_df.preview.str.len() > 1000] # 1000자가 넘는 데이터만 살려줌

preview_df
```

1506개에서 1383개 데이터로 줄어들었음

# 제목 column : [대여] 같은거 제거, 연속된 공백 제거, 중복제거

```python
title_preview_df = preview_df # 복사본
```

title_preview_df : 미리보기에다가 '제목'까지 전처리 된 데이터 프레임

```python
title_preview_df.title = title_preview_df.title.str.replace(r"\[.+\]", "", regex=True) # [ ] 제거

title_preview_df.title = title_preview_df.title.str.replace(r"\s+", " ", regex=True)

title_preview_df.title = title_preview_df.title.str.strip()

title_preview_df
```

제목이 같은 데이터를 제거했더니 1383개에서 1308개로 줄어들었음  
근데 책 제목이 같은 거는 지우면 안되고 하나만 남겨야 되는데 ㅋㅋ

## 인덱스 다시 맞춰주자

```python
title_preview_df = title_preview_df.reset_index()

title_preview_df
```

drop = True를 안해주면 이 꼴 남

```python
title_preview_df = title_preview_df.reset_index(drop=True)

title_preview_df
```

소용없음 그래서 다시 복사본 만들자 (백업느낌)

```python
title_preview_df = preview_df # 복사본

title_preview_df.title = title_preview_df.title.str.replace(r"\[.+\]", "", regex=True) # [ ] 제거
title_preview_df.title = title_preview_df.title.str.replace(r"\s+", " ", regex=True)
title_preview_df.title = title_preview_df.title.str.strip()
title_preview_df = title_preview_df.drop_duplicates(['title']) # 중복된 데이터 제거
title_preview_df = title_preview_df.drop_duplicates(["preview"]) # 중복된 데이터 제거
title_preview_df
```

중복된 내용을 제거하고 나니 1308개에서 1285개로 줄어들음

```python
title_preview_df = title_preview_df.reset_index(drop=True)

title_preview_df
```

```python
final_df = title_preview_df 
```

final_df : 전처리 마친 데이터 프레임
